```python
# Importing the necessary packages
import sys
import pandas as pd
import pymongo
import json
import os


# Defining tables of the database
def import_content1(filepath1):
    mng_client = pymongo.MongoClient('localhost', 27017)
    mng_db = mng_client['Assignment-2']
    collection_name = 'News' 
    db_cm = mng_db[collection_name]
    cdir = os.path.dirname('__file__')
    file_res = os.path.join(cdir, filepath1)
    
    fields = ['Time','Author','Title','News Content']
    data = pd.read_csv(file_res, usecols = fields)
    data_json = json.loads(data.to_json(orient='records'))
    db_cm.remove()
    db_cm.insert(data_json)
    
def import_content2(filepath2):
    mng_client = pymongo.MongoClient('localhost', 27017)
    mng_db = mng_client['Assignment-2'] 
    collection_name = 'Twitter' 
    db_cm = mng_db[collection_name]
    cdir = os.path.dirname('__file__')
    file_res = os.path.join(cdir, filepath2)
    
    fields = ['Time','Username','Location','Tweet Content']
    data = pd.read_csv(file_res, usecols = fields)
    data_json = json.loads(data.to_json(orient='records'))
    db_cm.remove()
    db_cm.insert(data_json)

def import_content3(filepath3):
    mng_client = pymongo.MongoClient('localhost', 27017)
    mng_db = mng_client['Assignment-2']
    collection_name = 'Clean-News' 
    db_cm = mng_db[collection_name]
    cdir = os.path.dirname('__file__')
    file_res = os.path.join(cdir, filepath3)
    
    fields = ['Time','Author','Title','News Content']
    data = pd.read_csv(file_res, usecols = fields)
    data_json = json.loads(data.to_json(orient='records'))
    db_cm.remove()
    db_cm.insert(data_json)

def import_content4(filepath4):
    mng_client = pymongo.MongoClient('localhost', 27017)
    mng_db = mng_client['Assignment-2'] 
    collection_name = 'Clean-Twitter' 
    db_cm = mng_db[collection_name]
    cdir = os.path.dirname('__file__')
    file_res = os.path.join(cdir, filepath4)
    
    fields = ['Time','Username','Location','Tweet Content']
    data = pd.read_csv(file_res, usecols = fields)
    data_json = json.loads(data.to_json(orient='records'))
    db_cm.remove()
    db_cm.insert(data_json)

    
# Importing data from the csv files
filepath1 = '/Users/admin/Desktop/Assignment-2 DWM/News.csv' 
import_content1(filepath1)

filepath2 = '/Users/admin/Desktop/Assignment-2 DWM/Twitter.csv' 
import_content2(filepath2)

filepath3 = '/Users/admin/Desktop/Assignment-2 DWM/Clean-News.csv' 
import_content3(filepath3)

filepath4 = '/Users/admin/Desktop/Assignment-2 DWM/Clean-Twitter.csv' 
import_content4(filepath4)
```
