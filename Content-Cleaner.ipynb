```python
# Importing the necessary packages
import re
import pandas
import os
import csv

############################################## Twitter #######################################################

# Defining and opening a csv file to store the data
csvfile = open("Clean-Twitter.csv", 'wt', newline='', encoding="utf-8")
writer = csv.writer(csvfile)


# Defining the headers inside the csv file
csvrow = []
csvrow.append("Time")
csvrow.append("Username")
csvrow.append("Location")
csvrow.append("Tweet Content")
writer.writerow(csvrow)


# Defining the cleaning function
def cleaner(string):
    string1 = re.sub('[^@a-zA-Z0-9:]', ' ', string)
    result = re.sub(r"http\S+", "", string1)
    return result

        
# Reading the unformatted file and inserting clean data into the new csv file   
with open('Twitter-outside.csv', encoding='utf-8') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        csvrow = []
        
        if line_count > 0:
            csvrow.append(row[0])
            string = row[1]
            csvrow.append(cleaner(string))
            string = row[2]
            csvrow.append(cleaner(string))
            string = row[3]
            csvrow.append(cleaner(string))
            writer.writerow(csvrow)
            line_count += 1
        else:
            line_count += 1
            
            
# Closing the csv file
csvfile.close()


# Indicator to know that extracting and storing is finished successfully
print("Done")


################################################ News #########################################################

# Defining and opening a csv file to store the data
csvfile = open("Clean-News.csv", 'wt', newline='', encoding="utf-8")
writer = csv.writer(csvfile)


# Defining the headers inside the csv file
csvrow = []
csvrow.append("Title")
csvrow.append("Author")
csvrow.append("Time")
csvrow.append("News Content")
writer.writerow(csvrow)


# Defining the cleaning function
def cleaner(string):
    string1 = re.sub('[^@a-zA-Z0-9:]', ' ', string)
    result = re.sub(r"http\S+", "", string1)
    return result

        
# Reading the unformatted file and inserting clean data into the new csv file   
with open('News-outside.csv', encoding='utf-8') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        csvrow = []
        
        if line_count > 0:
            string = row[0]
            csvrow.append(cleaner(string))
            string = row[1]
            csvrow.append(cleaner(string))
            csvrow.append(row[2])
            string = row[3]
            csvrow.append(cleaner(string))
            writer.writerow(csvrow)
            line_count += 1
        else:
            line_count += 1
            

# Closing the csv file
csvfile.close()


# Indicator to know that extracting and storing is finished successfully
print("Done")
```
